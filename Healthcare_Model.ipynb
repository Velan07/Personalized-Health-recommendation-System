##Dependent and Independent data split

```python
X = data.drop('prognosis', axis=1)
y = data['prognosis']
```

##LabelEncoder for Target variables

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y_label_dict = dict(zip(le.transform(le.classes_), le.classes_))
Y_label_dict
```

##Data spliting into training and testing in 7:3 ratio

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

##Model Building

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
models = {
    'SVC':SVC(kernel='linear'),
    'KNN':KNeighborsClassifier(),
    'RF':RandomForestClassifier(),
    'NB':GaussianNB()
}
```

##Model Evaluation

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f'Accuracy Score of the {name} model is, ', accuracy_score(y_test, y_pred))
    print(f'Confusion Matrix of the {name} model is, \n', confusion_matrix(y_test, y_pred))
    print('*'*100)
```

##Single Model Evaluation

```python
svc = SVC(kernel='linear')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(f'Accuracy Score of the SVC model is, ', accuracy_score(y_test, y_pred))
```

##Creating Dictionary for symptoms and Diseases

```python
symptoms_columns = np.asarray(data.columns)
symptoms_dict = dict()
for index, column in enumerate(symptoms_columns):
    symptoms_dict[column] = index
```

```python
symptoms_dict.pop('prognosis')
```

###Symptoms Dictionary

```python
symptoms_dict
```

###Disease Dictionary

```python
Y_label_dict = dict(zip(le.transform(le.classes_), le.classes_))
Y_label_dict
```

##Disease Name Prediction

```python
def get_perdicted_disease(input_symptoms):
    input_data = np.zeros(len(symptoms_dict))
    for items in input_symptoms:
        input_data[symptoms_dict[items]] = 1
    input_data = input_data.reshape(1, -1)
    y_pred = svc.predict(input_data)
    return Y_label_dict[y_pred[0]]
```

```python
input_symptoms = ['itching', 'skin_rash', 'nodal_skin_eruptions']
predicted_disease = get_perdicted_disease(input_symptoms)
print(f'Predicted disease: {predicted_disease}')
```

##Reading All Other Attributes for Disease

```python
symptoms_severity = pd.read_csv('D:\Project\Dataset\Symptom-severity.csv')
description = pd.read_csv('D:\Project\Dataset\description.csv')
precaution = pd.read_csv('D:\Project\Dataset\precautions.csv')
diets = pd.read_csv('D:\Project\Dataset\diets.csv')
symptoms = pd.read_csv('D:\Project\Dataset\symptoms.csv')
workouts = pd.read_csv('D:\Project\Dataset\workouts.csv')
medications = pd.read_csv('D:\Project\Dataset\medications.csv')
```

##Functions for getting Attributes

```python
def get_description(disease):
    return description[description['Disease'] == disease]['Description'][0]

def get_diets(disease):
    return diets[diets['Disease'] == disease]['Diet'][0]

def get_precautions(disease):
    pre = precaution[precaution['Disease'] == disease][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']].values
    return pre[0]

def get_workouts(disease):
    return np.asarray(workouts[workouts['disease'] == disease]['workout'])

def get_medications(disease):
    return medications[medications['Disease'] == disease]['Medication'][0]

def get_symptoms(disease):
    sym = symptoms[symptoms['Disease'] == disease][['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']].dropna()
    sym1 = sym.Symptom_1.unique().tolist()
    sym2 = sym.Symptom_2.unique().tolist()
    sym3 = sym.Symptom_3.unique().tolist()
    sym4 = sym.Symptom_4.unique().tolist()
    all_symptoms = sym1 + sym2 + sym3 + sym4
    return all_symptoms
```

```python
des = get_description(predicted_disease)
print('Description : ', list(des))
print('')
pre = get_precautions(predicted_disease)
print('Precautions : ', pre)
print('')
diet = get_diets(predicted_disease)
print('Diets : ', diet)
print('')
med = get_medications(predicted_disease)
print('Medications : ', med)
print('')
work = get_workouts(predicted_disease)
print('Workouts : ', work)
print('')
sym = get_symptoms(predicted_disease)
print('Common Symptoms : ', sym)
```
